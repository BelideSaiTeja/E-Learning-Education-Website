```python
sns.set_style('whitegrid')
plt.figure(figsize=(6,6))
plt.plot(data2021.Institute_Name, Performance.TLR)
plt.plot(data2021.Institute_Name, Performance.RPC)
plt.plot(data2021.Institute_Name, Performance.GO)
plt.plot(data2021.Institute_Name, Performance.OI)
plt.plot(data2021.Institute_Name, Performance.Perception)
plt.title('Institute vs Performance Factors')
plt.legend(['TLR', 'RPC', 'GO', 'OI', 'Perception'], bbox_to_anchor=(1.05, 1), loc='upper left');
```

```python
plt.rcParams['figure.figsize'] = [12,8]
k=['Rank','Score','TLR','RPC','GO','OI','Perception']
for i,j in data2021.groupby('year'):
    plt.title(f"Heatmap of Various Parameters for {i}")
    sns.heatmap(j[k].corr())
    plt.show()
```

```python
k=['Score','TLR','RPC','GO','OI','Perception']
for x in k:
    for i,j in data2021.groupby('year'):
        plt.plot(j['Rank'],j[f"{x}"],'s-',label=f"{i}")
    plt.title(f"Dependency of Rank on {x}")
    plt.ylabel(f"{x}")
    plt.xlabel("NIRF Rank")
    plt.legend()
    plt.show()
```

```python
plt.rcParams['figure.figsize'] = [12,8]

cols = ['Score','TLR','RPC','GO','OI','Perception']
for i,j in data2021.groupby("year"):
    plt.title(f"For the year {i}")
    j.boxplot(column=cols, fontsize=15)
    plt.show()
```
